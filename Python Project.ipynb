```python
print("-------------------------------------------------------------") 
print("Run Rate Calculator") 
print("-------------------------------------------------------------") 
firstInningsTotal = 0.00 
inGame = "y" 
print("-------------------------------------------------------------") 
print("Welcome to the Run Rate Calculator") 
print("-------------------------------------------------------------") 
while True: 
    try: 
        isFirstInnings = str.lower((input("Has your team played the First innings? (y/n) ")))  
    except ValueError: 
        print("Invalid please try again.") 
    if isFirstInnings not in ('y','n'): 
        print("Invalid please try again.") 
    else: 
        break  
gameLength = 0
while True: 
    try: 
        gameLength = int(input("Game length (in overs)? (20/50) "))  
    except ValueError: 
        print("Invalid please try again.") 
  
    if gameLength == 20: 
        break 
    elif gameLength == 50: 
        break  
    else: 
        print("Invalid please try again.") 
if isFirstInnings == "y": 
    while inGame == "y": 
        oversUsed = 0.00 
        while True: 
            try:  
                oversUsed = float(input("Number of Overs used? "))  
            except ValueError: 
                print("Invalid please try again.") 
            if oversUsed <= 0: 
                print("Invalid please try again.") 
            else: 
                break 
        currentRuns = 0.00 
        while True: 
            try: 
                currentRuns = float(input("Current Runs Scored? "))  
            except ValueError: 
                print("Invalid please try again.") 
            if currentRuns <= 0: 
                print("Invalid please try again.") 
            else: 
                break 
        runRate = currentRuns / oversUsed
        print("Current run rate ", "%.2f" % runRate, "\n") 
  
        while True: 
            try: 
                inGame = str.lower(input("Do you want to Update? (y/n) "))  
            except ValueError: 
                print("Invalid please try again.") 
            if inGame not in ('y','n'): 
                print("Invalid please try again.") 
            else: 
                break  
    else: 
        print("Thanks for using the Run Rate Calculator!!!") 
else: 
    while inGame == "y": 
        if firstInningsTotal == 0.00: 
            while True: 
                try: 
                    firstInningsTotal = float(input("1st innings run total? "))  
                except ValueError: 
                    print("Invalid please try again.") 
                if firstInningsTotal <= 0: 
                    print("Invalid please try again.") 
                else: 
                    break  
        else: 
            oversUsed = 0.00  
            while True: 
                try:  
                    oversUsed = float(input("Overs completed in the second innings? "))  
                except ValueError: 
                    print("Invalid please try again.") 
                if oversUsed <= 0:
                    print("Invalid please try again.") 
                else: 
                    break 
            currentRuns = 0.00 
            while True: 
                try: 
                    currentRuns = float(input("Current run total? "))  
                except ValueError: 
                    print("Invalid please try again.") 
                if currentRuns <= 0: 
                    print("Invalid please try again.") 
                else: 
                    break 
            runDifferance = (firstInningsTotal - currentRuns) + 1 
            oversRemaining = gameLength - oversUsed 
            runRate = runDifferance / oversRemaining 
            print("Required run rate : ", "%.2f" % runRate, "\n") 
  
            while True: 
                try: 
                    inGame = str.lower(input("Do you want to Update? (y/n) "))  
                except ValueError: 
                    print("Invalid please try again.") 
                if inGame not in ('y','n'): 
                    print("Invalid please try again.") 
                else: 
                    break  
    else: 
        print("Thanks for using the Run Rate Calculator") 
print("-------------------------------------------------------------") 
```
